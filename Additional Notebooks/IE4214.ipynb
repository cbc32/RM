```python
# 2 products assortment: Maximum expected revenue of 1240.274043 is acheived with price 3264.432563
import matplotlib.pyplot as plt
import numpy as np


beta=[0.20018419311203764, 0.5584902269506639, -0.015420963964399029, -0.0004940324550843351]
x_a = [4, 3.2, 95] # from the problem description

a = np.dot(beta[:3], x_a)
b = beta[3]
print('Expected utility is %f%fp' % (a,b))

def R(p):
    # watch for overflow!
    return p/(np.exp(-a-b*p)+1)
    # return p*np.exp(a+b*p)/(np.exp(a+b*p)+1)
    # return p*expit(a+b*p)

# Plot revenue for n prices uniformly distributed between up and low.
up = 3500
low = 2500
n = 1000
price = np.array(range(0,n,1))/n*(up-low)+low
revenue = [R(p) for p in price]
# print(revenue)
plt.plot(price,revenue)
plt.xlabel("Price")
plt.ylabel("Revenue")
plt.show()
```

```python
# Implementation i
from scipy.optimize import fmin

p_star = fmin(lambda x: -R(x), 0)[0]

R_star = R(p_star)

f"Maximum expected revenue of %f is acheived with price %f" % (R_star, p_star)
```

```python
# Theorem (Wang 2012)
p_star1 = R_star-1/b
p_star1
```

```python
# Implementation ii
# I uploaded a screenshot to Teams where WolframAlpha.com found this for me.
# Produces overflow with large a. Works with smaller a.
from scipy.special import lambertw

p_star2 = (-lambertw(np.exp(a-1))-1)/b

float(p_star2)
```

```python
def z(S):
    return np.dot(w[S],v[S])/(np.sum(v[S])+1)
```

```python
from itertools import combinations

z_list = []
S_list = []
Sz_list = []
for S in combinations(range(0,6),3):
    S = list(S)
    S_chr = [chr(ord('a')+n) for n in S]
    S_list.append(S_chr)
    z_list.append(z(S))
    Sz_list.append([S_chr,z(S)])

Sz_list
```
