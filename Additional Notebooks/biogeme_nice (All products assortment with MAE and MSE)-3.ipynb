```python
import numpy as np
import pandas as pd
import biogeme_nice ###my code
np.random.seed(42)
```

```python
df = pd.read_csv("clean_train.csv")

df["Assortment"] = df["Assortment"].apply(lambda x: [int(i) for i in x[1:-1].split(",")])
df["Probability"] = df["Probability"].apply(lambda x: [float(i) for i in x[1:-1].split(",")])
df
```

```python
### feature matrix given in problem instructions
ind = range(1,31)
x1 = [0] + [8 if 6<=i and i<=15 else 4 for i in ind]
x2 = [0] + [2.9 if 6<=i and i<=20 else 2.2 if 26<=i else 3.2 for i in ind]
x3 = [0] + [95 if i<=5 or (11<=i and i<=15) and i<=20 else 135 if 26<=i else 60 for i in ind]
x4 = [0] + [3000-300*((i-1)%5) for i in ind]
x = np.array([x1,x2,x3,x4]).T
```

```python
av_df = pd.read_csv("av_df100.csv")
av_df
```

```python
split = int(0.8*len(av_df))
train = av_df[:split]
```

```python
b_dict = biogeme_nice.calc_beta(train, x)
b_dict
```

```python
split = int(0.8*len(df))
test = df[split:].reset_index(drop=True)
pred =biogeme_nice.pred_df(b_dict, x, test["Assortment"])
test["Pred_Probability"] = pred["Pred"]
test
```

```python
def mae(y,q):
    return np.mean(np.abs(np.subtract(y,q)))
```

```python
mae_list = []
for i in range(len(test)):
    mae_list.append(mae(test["Probability"][i], test["Pred_Probability"][i]))
f'Mean of MAE (without intercept): %f' % np.mean(mae_list)
```

```python
b_dict = biogeme_nice.calc_beta(train, x, intercept=True) #the function will estimate an intercept term in addition to the other beta parameters

test1 = df[split:].reset_index(drop=True)
pred = biogeme_nice.pred_df(b_dict, x, test1["Assortment"])
test1["Pred_Probability"] = pred["Pred"]

mae_list = []
for i in range(len(test)):
    mae_list.append(mae(test1["Probability"][i], test1["Pred_Probability"][i]))
f'Mean of MAE (with intercept): %f' % np.mean(mae_list)
```

```python
from sklearn.metrics import mean_squared_error as mse
mse_list = []
for i in range(len(test)):
    mse_list.append(mse(test["Probability"][i], test["Pred_Probability"][i]))
f'Mean of MSE (without intercept): %f' % np.mean(mse_list)
```

```python
f'Mean of RMSE (without intercept): %f' % np.sqrt(np.mean(mse_list))
```
